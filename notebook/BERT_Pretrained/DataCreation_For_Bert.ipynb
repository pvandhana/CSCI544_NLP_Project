```python
def add11(a):
    result = model(tokenizer.encode(a, return_tensors='pt', truncation=True, max_length=512))
    xt = int(torch.argmax(result.logits))+1
    return xt
```

# Normalizing likes for each comment for a video_id

```python
for i in range(0,36963):
    mi=2500000
    ma=-1
    for j in range(1,21):
        if(vc['comment_{}_likes'.format(j)][i] > ma):
            ma = vc['comment_{}_likes'.format(j)][i]
        if(vc['comment_{}_likes'.format(j)][i] < mi):
            mi = vc['comment_{}_likes'.format(j)][i]
    for j in range(1,21):
        if(ma == mi):
            vc['comment_{}_n'.format(j)][i] = 1
        else:
            vc['comment_{}_n'.format(j)][i] = (vc['comment_{}_likes'.format(j)][i]-mi)/(ma-mi)
```
